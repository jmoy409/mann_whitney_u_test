```python
import PyNonpar
from PyNonpar import*

question = 'Column of interest in Pandas dataframe'
ref = opt_in_df.loc[(~np.isnan(opt_in_df[question]))]
#create synthetic representation of column and subtract .5(desired effect size) from values
#effect size we are creating is .5, suggesting there is a .5 difference bettween both datasets
syn = ref[question] - 0.5
#t=1/2 # is nummber of cases in reference and synthetic data
rst = PyNonpar.twosample.wilcoxon_mann_whitney_ssp(ref[question].tolist(), syn.tolist(), power=0.9, alpha=0.05, t=1/2)
#sample size needed for each sample group to perform mann whitney U test for desired effect size, power, and alpha
#ex) 150 means need 150 samples for df1 and df2 = 300 total
sample_size = int(round(rst.n1,0))

#Perform again with # of samples needed
df1_sample = df1.sample(sample_size)
df2_sample = df2.sample(sample_size)
stats.mannwhitneyu(df1_sample[question], df2_sample[question], alternative='two-sided')
```

```python
#Function to run through multiple columns

def mann_u_test(df1,df2, columns):
    np.random.seed(2023) 
    for question in columns:
        ref = opt_in_df.loc[(~np.isnan(opt_in_df[question]))]
        syn = ref[question] - 0.5
        rst = PyNonpar.twosample.wilcoxon_mann_whitney_ssp(ref[question].tolist(), syn.tolist(), power=0.9, alpha=0.05, t=1/2)
        sample_size = int(round(rst.n1,0))
        df1_sample = df1.sample(sample_size)
        df2_sample = df2.sample(sample_size)
        print('question: ', question)
        print('sample size: ', question)
        result = stats.mannwhitneyu(df1_sample[question], df2_sample[question], alternative='two-sided')
        print(result.pvalue)
```
